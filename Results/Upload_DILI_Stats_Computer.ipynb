```python
import pandas as pd
results_df = pd.read_csv("/content/drive/MyDrive/ClinicalTrials/uniTox Ratings/FinaluniTox/Sunday_DILI_direct_4o_livertoxicity_final.csv")
```

```python
less_dilirank_df = pd.read_csv("dilirank_df_spl.csv")
less_dilirank_df = less_dilirank_df.merge(results_df,left_on=["SPL"],right_on=["set_id"],how="inner")
less_dilirank_df = less_dilirank_df.groupby("LabelCompoundName").first().reset_index()
```

```python
less_dilirank_df
```

```python
less_dilirank_df["Answer"] = less_dilirank_df["vDILIConcern"].map({"Ambiguous DILI-concern":1,"vNo-DILI-Concern":0,"vMost-DILI-Concern":1,"vLess-DILI-Concern":1})
less_dilirank_df["Pred"] = less_dilirank_df["less_determination"].map({"No":0,"No.":0,"Less":1,"Most":1})
```

```python
print(less_dilirank_df[less_dilirank_df["Answer"].isna()])
print(less_dilirank_df[less_dilirank_df["Pred"].isna()])
```

```python
print((less_dilirank_df["Answer"]==less_dilirank_df["Pred"]).sum()/len(less_dilirank_df))
```

```python
len(less_dilirank_df[(less_dilirank_df["Answer"]==0) & (less_dilirank_df["Pred"]==0)])/len(less_dilirank_df[less_dilirank_df["Pred"]==0])
```

```python
len(less_dilirank_df[(less_dilirank_df["Answer"]==1) & (less_dilirank_df["Pred"]==1)])/len(less_dilirank_df[less_dilirank_df["Pred"]==1])
```

```python
less_dilirank_df = less_dilirank_df[less_dilirank_df["less_determination"].isin(["No","Most"])]
```

```python
len(less_dilirank_df[(less_dilirank_df["Answer"]==1) & (less_dilirank_df["Pred"]==1)])/len(less_dilirank_df[less_dilirank_df["Pred"]==1])
```

```python
len(binary_dilirank_df[(binary_dilirank_df["Answer"]==1) & (binary_dilirank_df["Pred"]==1)])/len(binary_dilirank_df[binary_dilirank_df["Pred"]==1])
```
